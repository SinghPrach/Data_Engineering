```
def findAvgSimple(dfr,coln):  # define Average Value function
  average = dfr.agg({coln:'avg'})
  return average.collect()[0][0]
##returns double

def findMinValue(dfr,coln):  # define Minimum Value function
    MinVal = dfr.agg({coln:'min'})
    return MinVal.collect()[0][0]

def findMaxValue(dfr,coln):   # define Maximum Value function
    MaxVal = dfr.agg({coln:'max'})
    return MaxVal.collect()[0][0]

def findSumValue(dfr,coln):  # define Sum Value function
    SumVal = dfr.agg({coln:'sum'})
    return SumVal.collect()[0][0]

def findSDValue(dfr,coln):   # define Standard Deviation Value function
    SDVal = dfr.agg({coln:'stddev_pop'})
    return SDVal.collect()[0][0]
```

```
stringCols = []
nonStringCols = []
for coln in df1.dtypes:     #df.dtypes returns a list
  if coln[1] == 'string':
    stringCols.append(coln[0])
  else:
    nonStringCols.append(coln[0])

averages = []
mins = []
maxs = []
sums = []
SDs = []

for c in nonStringCols:
  averages.append(findAvgSimple(df1,c))
  mins.append(findMinValue(df1,c))
  maxs.append(findMaxValue(df1,c))
  sums.append(findSumValue(df1,c))
  SDs.append(findSDValue(df1,c))

AverageDict = dict(zip(nonStringCols, averages))
MinDict = dict(zip(nonStringCols, mins))
MaxDict = dict(zip(nonStringCols, maxs))
SumDict = dict(zip(nonStringCols, sums))
SDDict = dict(zip(nonStringCols, SDs))
#print(nonStringCols)
#print(averages)
# print("Averages of the Column")
# print(AverageDict)
# print("Minimum Values in the Column")
# print(MinDict)
# print("Maximum Values in the Column")
# print(MaxDict)
# print("Sum Values in the Column")
# print(SumDict)
# print("Standard Deviation Values in the Column")
# print(SDDict)
```
