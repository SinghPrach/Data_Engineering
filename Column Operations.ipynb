```
# Import Spark libraries
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType,StructField, StringType, IntegerType, DecimalType
from pyspark.sql.functions import col
from pyspark.sql.functions import split,lit
```

```
df_sample_data = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/27prachisingh@gmail.com/Employee_Sample_Data.csv")
```

```
#df_sample_data.show()
#Using withColumn()
#1. Add a new column
#df_new_col=df_sample_data.withColumn("salary",split(df_sample_data['Annual Salary'], ',').getItem(0))
#df_new_col.show()
df_new_col=df_sample_data.withColumn("number",split(df_sample_data['Bonus %'], '%').getItem(0))
df_new_col.show()
```

```
#Change DataType using PySpark withColumn()
df_new_col=df_new_col.withColumn("number",col("number").cast("Integer"))
#Special Casting
#targetDf.withColumn('totalValueCR',targetDf.totalValueCR.cast(DecimalType(32,3)))
```

```
# Update The Value of an Existing Column
df_new_col.withColumn("number",col("number")*100).show()
```

```
#Add a New Column using withColumn()
df_new_col.withColumn("Country1", lit("USA")).show()
df_new_col.withColumn("Country1", lit("USA")) \
  .withColumn("anotherColumn",lit("anotherValue")) \
  .show()
```

```
#Rename a column
df_new_col.withColumnRenamed("number","new_col") \
  .show(truncate=False) 
df_new_col=df_new_col.withColumnRenamed("Exit Date","ExitDate")
```

```
#Drop a column
#df.drop("salary") \
 # .show() 
```

```
#Filtering
# Using equals condition
df_new_col.filter(df_new_col.City == "Chicago").show(truncate=False)
```

```
#df_new_col.filter(df.City != "Chicago") \
 #   .show(truncate=False) 
# df_new_col.filter(~(df.City == "Chicago")) \
#    .show(truncate=False)
df_new_col.filter(df_new_col.number>7).show()
```

```
#Filter multiple condition
df_new_col.filter( (df_new_col.City  == "Chicago") & (df_new_col.Gender  == "Male") ) \
    .show(truncate=False)  
```

```
#Filter IS IN List values
li=["Chicago","Seattle","Austin"]
df_new_col.filter(df_new_col.City.isin(li)).show()
```

```
#Filter Based on Starts With, Ends With, Contains
#Starts With
df_new_col.filter(df_new_col.City.startswith("A")).show()
```

```
#Ends With
df_new_col.filter(df_new_col.City.endswith("O")).show()
```

```
#contains
df_new_col.filter(df_new_col.City.contains("cago")).show()
```

```
#PySpark Filter like and rlike
df_new_col.filter(df_new_col.City.like("%cago%")).show()
```

```
#Grouping
#groupBy(col1 : scala.Predef.String, cols : scala.Predef.String*) :
#     org.apache.spark.sql.RelationalGroupedDataset
# count() – Use groupBy() count() to return the number of rows for each group.
# mean() – Returns the mean of values for each group.
# max() – Returns the maximum of values for each group.
# min() – Returns the minimum of values for each group.
# sum() – Returns the total for values for each group.
# avg() – Returns the average for values for each group.
# agg() – Using groupBy() agg() function, we can calculate more than one aggregate at a time.
# pivot() – This function is used to Pivot the DataFrame
df_new_col.groupBy("Department").sum("number").show(truncate=False)
#df_new_col.printSchema()
# df_new_col.groupBy("Department").count()

# df_new_col.groupBy("Department").min("number")

# df_new_col.groupBy("Department").max("number")

# df_new_col.groupBy("Department").avg( "number")

# df_new_col.groupBy("Department").mean( "number") 
```

```
df_new_col.groupBy("Department","City") \
    .avg("number") \
    .show(truncate=False)
```
